```python
import pandas as pd
import random
from faker import Faker
```

```python
fake = Faker()
```

```python
total_p = 50
product_c = ["Electronics", "Apparel", "Home Appliances", "Books", "Toys", "Groceries"]

p = []
for i in range(1, total_p + 1):
    p.append({"Pdt_id": f"P{str(i).zfill(3)}",
               "Pdt_c": random.choice(product_c)})
```

```python
data = []
users = 1000

for i in range(1, users + 1):
    user_id = f"U{str(i).zfill(4)}"
    num_purchases = random.randint(1, 5)
    purchase_history = []

    for _ in range(num_purchases):
        p1 = random.choice(p)
        purchase_date = fake.date_this_year()
        purchase_history.append((p1["Pdt_id"], purchase_date))

        data.append({
            "User _ID": user_id,
            "Product_ID": p1["Pdt_id"],
            "Product_Category": p1["Pdt_c"],
            "Purchase History": purchase_history,
            "Rating": round(random.uniform(1.0, 5.0), 1)
        })
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv("Mock_data.csv",index=False)
```
